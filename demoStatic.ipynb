```python
import cv2
import numpy as np
import os
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  
import tensorflow as tf
from yolov3.yolov4 import Create_Yolo
from yolov3.utils import load_yolo_weights, detect_image
from yolov3.configs import *

if YOLO_TYPE == "yolov4":
    Darknet_weights = YOLO_V4_TINY_WEIGHTS if TRAIN_YOLO_TINY else YOLO_V4_WEIGHTS
if YOLO_TYPE == "yolov3":
    Darknet_weights = YOLO_V3_TINY_WEIGHTS if TRAIN_YOLO_TINY else YOLO_V3_WEIGHTS

yolo = Create_Yolo(input_size=YOLO_INPUT_SIZE)
load_yolo_weights(yolo, Darknet_weights) # use Darknet weights
```

```python
from gtts import gTTS
import os
```

```python
def getLoc(x,y,width,height):
    loc = {
        "top left":((0,width/3),(0,height/3)),
        "top":((width/3,(2*width)/3),(0,height/3)),
        "top right":(((2*width)/3,width),(0,height/3)),
        "middle left":((0,width/3),(height/3,(2*height)/3)),
        "middle":((width/3,(2*width)/3),(height/3,(2*height)/3)),
        "middle right":(((2*width)/3,width),(height/3,(2*height)/3)),
        "bottom left":((0,width/3),((2*height)/3,height)),
        "bottom":((width/3,(2*width)/3),((2*height)/3,height)),
        "bottom right":(((2*width)/3,width),((2*height)/3,height)),
    }
    pos = ""
    for key in loc:
        xr = loc[key][0]
        xy = loc[key][1]
        if (x >= xr[0] and x<xr[1]) and (y>=xy[0] and y<xy[1]):
            pos = key
            
    return pos
```

```python
image_path   = "./IMAGES/bottle.jpeg"

image,cl = detect_image(yolo, image_path, '', input_size=YOLO_INPUT_SIZE, show=False, rectangle_colors=(255,0,0))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#print(cl["bottle"])

height, width = image.shape[:2]
#print(height)
#print(width)

input_s = ""

plt.figure(figsize=(5,15))
plt.imshow(image)

for key in cl:
    #print(type(key))
    for y in cl[key]:
        x = (y[0] + y[2])/2
        y = (y[1] + y[3])/2
        ans = getLoc(x,y,width,height)
        input_s = input_s + key + " at " + ans + " "
#         status = safe()
        #print(ans)
        #print(type(ans))
        plt.scatter(x,y)
plt.show()
print(input_s)

```
